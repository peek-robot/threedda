```python
import os
import json

root_dir = "/lustre/fs12/portfolios/nvr/users/mmemmel/projects/vila/data/oxe_processed_path_mask_subtraj"

code_snippets = []
dataset_names = []
for folder in sorted(os.listdir(root_dir)):
    folder_path = os.path.join(root_dir, folder)
    if os.path.isdir(folder_path):
        json_files = [f for f in os.listdir(folder_path) if f.endswith('.json')]
        if json_files:
            dataset_name = folder
            data_path = os.path.join(folder_path, json_files[0])
            image_path = os.path.join(folder_path, "images")
            snippet = f"""{dataset_name} = Dataset(
    dataset_name='{dataset_name}',
    dataset_type='torch',
    data_path='{data_path}',
    image_path='{image_path}'
)
add_dataset({dataset_name})
"""
            code_snippets.append(snippet)
            dataset_names.append(dataset_name)

for snippet in code_snippets:
    print(snippet)

joined_dataset_names = "+".join(dataset_names)
print("datasets", len(code_snippets))
print("dataset_mixture", joined_dataset_names)
```
